Remove WaterMark

```python
from nowatermark . nowatermark import  WatermarkRemover

input_ = "example.jpg"

output = "output.jpg"

mask = "mask.png"

remover = WatermarkRemover ( )
remover.load_watermark_template( mask )

remover.remove_watermark( input_ , output )

print("Complete")
```
